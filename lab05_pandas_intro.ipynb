## Ćwiczenie 1
Series — tworzenie i operacje

```python
import pandas as pd
import numpy as np
```

```python
oceny = pd.Series([4.5, 3.0, 5.0, 3.5, 4.0])
print(oceny)
```

```python
pensje = {'Anna': 5500, 'Jan': 7200, 'Ewa': 4800, 'Marek': 9100, 'Kasia': 6300}
print(pensje)
pensje = pd.Series(pensje)
print(pensje)

imie = pensje.idxmax()
print(f'Najwięcej zarabia: {imie}')
print(f'Średnia pensja: {pensje.mean()}')
maska = (pensje > pensje.mean())
print(f'Osoby zarabiające poniżej średniej: \n{pensje[maska]}')
print(f'Poniżej 6000 zarabia {(pensje < 6000).sum()} osób')
```

```python
podwyzka = pensje * 1.10
print(podwyzka)
premia = pensje + 500
print(premia)
nowePensje = (pensje * 1.10) + 500
print(nowePensje)
print(pensje.sort_values(ascending=False))
```

## Ćwiczenie 4
 Pytania biznesowe

```python
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
tips['tip_pct'] = (tips['tip'] / tips['total_bill'] * 100).round(1)
```

```python
print(tips['total_bill'].mean())
print(tips['tip_pct'].mean())
print(tips['day'].value_counts())
print(tips['smoker'].value_counts())
print(tips['tip'].describe())
print(tips['time'].value_counts(normalize= True))
print(tips['size'].nunique())
```
