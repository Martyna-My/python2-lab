## Ćwiczenie 3
Operacje wektorowe

```python
nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka',
'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka']

ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])
```

Zadanie 1: Podstawowe statystyki cenowe

```python
print (f'Średnia cena: {ceny.mean()}')
id_min = ceny.argmin()
id_max = ceny.argmax()
print (f'Najtańszy produkt: {ceny.min()} - {nazwy[id_min]}')
print (f'Najdroższy produkt {ceny.max()} - {nazwy[id_max]}')
```

Zadanie 2: Wartość magazynu

```python
wartosci = ceny * ilosci_magazyn
print(f'Wartość każdego produktu: {wartosci}')
print(f'Łączna wartość magazynu: {wartosci.sum()}')
```

Zadanie 3: Analiza cenowa

```python
promo = ceny * 0.85
vat = ceny * 1.23
print(f'Ceny po rabacie: {promo}')
print(f'Ceny z VAT: {vat}')
print(f'Produkty powyżej 1000 zł: {(ceny > 1000).sum()}')
```

Zadanie 4: Filtrowanie

```python
goodrate = oceny_klientow >= 4.5
nazwy = np.array(nazwy)
print('Produkty z oceną >= 4.5:')
print(nazwy[goodrate])
print(ceny[goodrate])
```

```python
tanie = ceny < 200
print('Produkty poniżej 200 zł')
print(nazwy[tanie])
print(ilosci_magazyn[tanie])
```

Zadanie 5: Analiza zapasów

```python
niskistan = ilosci_magazyn < 30
wartosci = ceny * ilosci_magazyn
procent = (wartosci[niskistan].sum() / wartosci.sum()) * 100
print(f'Produkty z niskim stanem magazynowym: {nazwy[niskistan]}')
print(f'Łączna wartość tych produktów: {wartosci[niskistan].sum()}')
print(f'Procent magazynu: {procent}%')
```

Zadanie 6 (bonus): Analiza korelacji cena–ocena

```python
okazja = (oceny_klientow >= 4.5) & (ceny < 1000)
wysokie = (oceny_klientow >= 4.0)
niskie = (oceny_klientow < 4.0)
print(f'Produkty które są okazją: {nazwy[okazja]} - {ceny[okazja]}')
print(f'Średnia cena produktów z oceną >= 4.0 - {ceny[wysokie].mean()}, z oceną < 4.0 - {ceny[niskie].mean()}')
```

## Ćwiczenie 4
Benchmark - lista vs NumPy

```python
import time
```

Test 1: mnożenie

```python
rozmiar = 1_000_000
lista = list(range(rozmiar))
tablica = np.array(lista)
```

```python
start = time.perf_counter()
wynik_lista = [x * 2 for x in lista]
czas_lista = time.perf_counter()- start
```

```python
start = time.perf_counter()
wynik_numpy = tablica * 2
czas_numpy = time.perf_counter()- start
```

```python
print(f"=== Mnożenie ×2 ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

Test 2: suma

```python
start = time.perf_counter()
suma_lista = sum(lista)
czas_lista = time.perf_counter()- start
```

```python
start = time.perf_counter()
suma_numpy = tablica.sum()
czas_numpy = time.perf_counter()- start
```

```python
print(f"\n=== Suma ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

Test 3: flitrowanie (elementy > 500000)

```python
start = time.perf_counter()
filtr_lista = [x for x in lista if x > 500000]
czas_lista = time.perf_counter()- start
```

```python
start = time.perf_counter()
filtr_numpy = tablica[tablica > 500000]
czas_numpy = time.perf_counter()- start
```

```python
print(f"\n=== Filtrowanie > 500k ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

## Wnioski z benchmarku
- NumPy jest **14x** szybszy w mnożeniu
- NumPy jest **14×** szybszy w sumowaniu
- NumPy jest **9×** szybszy w filtrowaniu
- Wniosek: przy dużych danych **zawsze** używaj NumPy zamiast list
