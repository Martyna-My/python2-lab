### Ćwiczenie 3
Pełny pipeline czyszczenia

```python
df_clean = pd.DataFrame(data)

df_clean = df_clean.replace('brak', np.nan)
df_clean['wynagrodzenie'] = pd.to_numeric(df_clean['wynagrodzenie'], errors='coerce')

df_clean = df_clean.drop_duplicates().reset_index(drop=True)

median_salary = df_clean['wynagrodzenie'].median()
df_clean['wynagrodzenie'] = df_clean['wynagrodzenie'].fillna(median_salary)

mean_rating = round(df_clean['ocena_roczna'].mean(), 2)
df_clean['ocena_roczna'] = df_clean['ocena_roczna'].fillna(mean_rating)

df_clean['imie'] = df_clean['imie'].str.strip().str.title()

df_clean['dzial'] = (
    df_clean['dzial']
    .str.strip()
    .str.title()
    .str.replace('Hr', 'HR', regex=False)
    .str.replace('It', 'IT', regex=False)
)

df_clean['data_zatrudnienia'] = pd.to_datetime(df_clean['data_zatrudnienia'], errors='coerce')

print("=== WERYFIKACJA ===")
print(f"Shape: {df_clean.shape}")
print(f"\nDtypes:\n{df_clean.dtypes}")
print(f"\nNaN per kolumna:\n{df_clean.isna().sum()}")
print(f"\nUnikalne działy: {sorted(df_clean['dzial'].unique())}")
print(f"\nPierwsze imiona: {df_clean['imie'].tolist()[:8]}")
```

```python
print(f'Średnie wynagrodzenie per dział: {
     df_clean.groupby('dzial')['wynagrodzenie'].mean().round(2)
}')

print(f'Średnia ocena per dział: {
    df_clean.groupby('dzial')['ocena_roczna'].mean().round(2)
}')

print(f'Top 3 najlepiej opłacanych: {
    df_clean.nlargest(3, 'wynagrodzenie')[['imie', 'dzial', 'wynagrodzenie']]
}')

print(f'Liczba pracowników per dział: {
    df_clean['dzial'].value_counts()
}')
```

```python
dzisiaj = pd.Timestamp.today()
print(f'Najłuższy staż pracy: {
    round((dzisiaj - df_clean['data_zatrudnienia'].min()).days / 365, 2)
}')

```

```python
df_str['imie'] = df_str['imie'].str.strip().str.title()

df_str['dzial'] = (
    df_str['dzial']
    .str.strip()
    .str.title()
    .str.replace('Hr', 'HR', regex=False)
    .str.replace('It', 'IT', regex=False)
)

print(f'Imiona na A: {df_str[df_str['imie'].str.startswith('A')]}')

df_str['dlugosc_imienia'] = df_str['imie'].str.len()
print(df_str['dlugosc_imienia'].describe())

print(f'Najdłuższe imie: {df_str.nlargest(1, 'dlugosc_imienia')}')
```

```python
print(f'Pracownicy IT lub HR: {df_str[df_str['dzial'].str.contains('IT|HR')]}')

print(f'Imiona z literą A lub a: {df_str[df_str['imie'].str.contains('a', case=False)]}')

print(f'Imiona kończące się na a: {df_str['imie'].str.endswith('a').sum()}')
```
