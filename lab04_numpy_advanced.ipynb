## Ćwiczenie 2

## Ćwiczenie 3
Analiza danych finansowych

```python
np.random.seed(2026)

# 8 pracowników, dane roczne
pracownicy = ['Anna', 'Jan', 'Ewa', 'Marek', 'Kasia', 'Piotr', 'Zofia', 'Tomek']
pensje = np.array([5500, 7200, 4800, 9100, 6300, 4200, 8500, 5800])
staz = np.array([3, 8, 2, 15, 5, 1, 12, 4])
oceny = np.array([4.2, 3.8, 4.5, 4.0, 4.7, 3.5, 4.3, 3.9])

# Sprzedaż miesięczna 4 produktów przez 6 miesięcy
produkty_nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Akcesorium']
sprzedaz = np.random.randint(20, 200, size=(4, 6))
ceny_produktow = np.array([3500, 1800, 2500, 150])
```

```python
#1
print(f'Srednia pensja: {pensje.mean()}')
print(f'Mediana pensji: {np.median(pensje)}')
print(f'Odchylenie standardowe pensji: {pensje.std()}')
#2
pracownicy = np.array(pracownicy)
bogacze = (pensje > pensje.mean())
print(f'Osoby zarabiające powyżej sredniej: {pracownicy[bogacze]}')
#3
stazysci = (pensje < np.median(pensje))
print(f'Osoby zarabiające poniżej mediany: {pracownicy[stazysci]}')
#4
rozstep = pensje.max() - pensje.min()
iqr = np.percentile(pensje, 75) - np.percentile(pensje, 25)
print(f'Rozstęp wynagrodzeń: {rozstep}')
print(f'Rozstęp międzykwartalny: {iqr}')
```

```python
premia = np.where(oceny >= 4.5, 2000, np.where(oceny >= 4, 1000, 0))
noweWynagrodzenie = pensje + premia
print(noweWynagrodzenie)
idmax = premia.argmax()
print(pracownicy[idmax])
```

```python
kolejnosc = np.argsort(pensje)[::-1]
for i, idx in enumerate(kolejnosc, 1):
    print(i, pracownicy[idx], pensje[idx], staz[idx])
    i += 1
```

```python
korelacjaSP = np.corrcoef(staz, pensje)[0, 1]
korelacjaSO = np.corrcoef(staz, oceny)[0, 1]
print(f'Korelacja: staż vs pensja: {korelacjaSP}')
print(f'Korelacja: staż vs ocena: {korelacjaSO}')
print(f'Silniejszy związek jest między stażem a pensją')
```

```python
ceny_kol = ceny_produktow.reshape(4, 1)
przychod = sprzedaz * ceny_kol
print(przychod)
przychodMax = przychod.sum(axis = 1).argmax()
najlepszyMiesiąc = przychod.sum(axis = 0).argmax()
początek = przychod[:, :3].sum()
koniec = przychod[:, 3:].sum()
print(f'Największy łączny przychód przyniósł produkt: {produkty_nazwy[przychodMax]}')
print(f'Najlepszy miesiąc pod kątem sprzedaży to: {najlepszyMiesiąc + 1}')
print(f'Suma pierwsze 3 m-ce: {początek}, suma ostatnie 3 m-ce: {koniec}')
```

## Ćwiczenie 4
Generowanie danych 

```python
np.random.seed(42)
```

```python
wiek = np.random.normal(loc = 35, scale = 10, size = 100)
wiek = np.clip(wiek, 18, None).astype(int)
wydatki = np.clip(np.random.normal(500, 200, 100), 0, None)
wizyty = np.random.poisson(5, 100)
```

```python
print(f'Wiek - srednia: {wiek.mean():.1f}, mediana: {np.median(wiek):.1f}, odchylenie standardowe: {wiek.std():.1f}, minimum: {wiek.min()}, maksimum: {wiek.max()}')
print(f'Wydatki - srednia: {wydatki.mean():.1f}, mediana: {np.median(wydatki):.1f}, odchylenie standardowe: {wydatki.std():.1f}, minimum: {wydatki.min():.1f}, maksimum: {wydatki.max():.1f}')
print(f'Wizyty - srednia: {wizyty.mean():.1f}, mediana: {np.median(wizyty):.1f}, odchylenie standardowe: {wizyty.std():.1f}, minimum: {wizyty.min()}, maksimum: {wizyty.max()}')
```

```python
print(f'Korelacja: wiek vs wydatki: {np.corrcoef(wiek, wydatki)[0, 1]:.2f}')
print(f'Korelacja: wizyty vs wydatki: {np.corrcoef(wizyty, wydatki)[0, 1]:.2f}')
segmentacja = np.where(wydatki > wydatki.mean() + wydatki.std(), 'WYSOKI', np.where(wydatki < wydatki.mean() - wydatki.std(), 'NISKI', 'SREDNI'))
print(segmentacja)

```

```python
nazwy_seg, liczniki = np.unique(segmentacja, return_counts=True)
for nazwa, ile in zip(nazwy_seg, liczniki):
    print(f'Segment {nazwa}: {ile} klientow')
```
